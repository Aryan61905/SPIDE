```python
r=str(df["Date"].head(1))
df["Home/Neutral"].head(1)
months= {
    "Jan":"01","Feb":"02","Mar":"03","Apr":"04","May":"05","Jun":"06","Jul":"07","Aug":"08","Sep":"09","Oct":"10","Nov":"11","Dec":"12"
}
#print(r)

#r=r.split(',')[1][4:7]
#print(r)
df['Token'] = df['Date'].apply(lambda x: f"{x[-4:]}{months[x.split(',')[1][1:4]]}{r.split(',')[1][5:7]}")+"0"+df["Home/Neutral"].apply(lambda x: x[:3].upper())

token = str(df['Token'].head(1))
print(token)
visitor = str(df["Visitor/Neutral"].head(1))
visitor=visitor.split()[1].upper()
token=token.split()[1]
print(token)
#print(visitor)
```

```python
result_df = df[df['Date'].str.contains('Oct')]
result_df
```

```python

token_dict = {
    "GOL":"GSW", "SAN": "SAS", "BRO": "BRK", "CHA": "CHO", "CLIPPERS":"LAC", "LAKERS":"LAL", "NEW":"NOP", "PELICANS": "NOP", "KNICKS": "NYK", "OKL": "OKC"
}

def modify_token(value):
    team_chars = value[0:3].upper()
    if team_chars in ["NEW", "LOS"]: 
        team_chars = token_dict[value.split(" ")[-1].upper()]
    elif team_chars in token_dict:
        team_chars = token_dict[team_chars]
    
    return team_chars
result_df["Home_Token"] = result_df["Home/Neutral"].apply(modify_token)
result_df["Visitor_Token"] = result_df["Visitor/Neutral"].apply(modify_token)
result_df['Token'] = result_df['Date'].apply(lambda x: f"{x[-4:]}{months[x.split(',')[1][1:4]]}{x.split(',')[1][5:7]}")+"0"+result_df["Home_Token"]+"VS"+result_df["Visitor_Token"]
result_df
```

```python
fileName = result_df["Token"][0]
print(fileName)

box_df = pd.read_csv(f"/Users/roy/Desktop/SPIDE/BoxScores/{fileName}.csv",header=None,skiprows= 1)
new_row = [np.nan, "Basic Box Score Stats"] + [np.nan] * (len(box_df.columns) - 2)

# Insert the new row at the beginning
box_df.loc[-1] = new_row
box_df.index = box_df.index + 1  # Shift the index to accommodate the new row
box_df = box_df.sort_index() 

box_df.head(60)
```

```python
current_header = None
current_dataframe = None
dataframes = {}
home=fileName[9:12]
away=fileName[-3:]
print(home,away)
header_titles=["G","Q1","Q2","H1","Q3","Q4","H2"]
ind=0
flag=away
# Iterate through rows
for index, row in box_df.iterrows():
   
    if 'Basic Box Score Stats' in row[1]:
        # If a new header is found, save the current dataframe (if any)
        # and create a new one with the new header
        if current_header is not None:
            dataframes[current_header] = current_dataframe
        if ind==len(header_titles):
            ind = 0
            flag = home 
        current_header = header_titles[ind]+" "+flag
        print(current_header)
        ind+=1
        current_dataframe = pd.DataFrame(columns=box_df.columns)
    else:
        # Append the current row to the current dataframe
        current_dataframe = current_dataframe.append(pd.Series(row, index=box_df.columns), ignore_index=True)

# Save the last dataframe
if current_header is not None:
    if current_header not in dataframes:
        dataframes[current_header] = current_dataframe
```
