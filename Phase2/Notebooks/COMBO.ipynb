```python
import pandas as pd
```

```python

df = pd.read_csv("/Users/roy/Desktop/SPIDE/Phase2/csvs/OverallPlayerStats.csv")
```

```python
df.columns

df = df[df.PTS != "PTS"]
df["TRB"] = df["TRB"].astype(float)
df["PTS"] = df["PTS"].astype(float)
df["AST"] = df["AST"].astype(float)

df["COMBO"] = (df["PTS"])+(df["AST"])+(df["TRB"])
df = df.sort_values(by=["COMBO"],ascending=False)
points_df = df.sort_values(by=["COMBO"],ascending=False)

df.columns
```

```python
grouped_teams = df.groupby('Tm')
columns_to_include = ['Tm' ,'Player','G','MP','3P','3PA','TRB', 'AST','PTS', 'COMBO']
# Create separate DataFrames for each team
team_dataframes = {}
for team_name, group_data in grouped_teams:
    team_dataframes[team_name] = group_data.loc[:,columns_to_include]

# Access each team's DataFrame using the team name
```
